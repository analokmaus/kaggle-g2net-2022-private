```python
import pandas as pd
```

```python
DATA = 'data/G2Net2022Blend.csv'
LEAK = 'data/leak.csv'
```

```python
df_leak = pd.read_csv(LEAK)
df_sub = pd.read_csv(DATA)
df_leak = pd.merge(df_leak,df_sub,on='id')
df_leak['target_adj'] = df_leak['target']
df_leak.head()
```

```python
# adjust positive-like examples
df_leak.target_adj = df_leak.target + (1 - df_leak.target)*\
    ((0.5*(df_leak.mask_avr_H + df_leak.mask_avr_L) - 0.01)/0.03).clip(0,1)

# adjust negative-like examples
df_leak.target_adj = df_leak.target_adj*(1 - ((df_leak.f - 0.65)/0.45).clip(0,1)*\
                   (((0.01 - 0.5*(df_leak.mask_avr_H + df_leak.mask_avr_L))/0.01).clip(0,1)))
```

```python
df_sub = df_sub.set_index('id')
df_sub.loc[df_leak.id,'target'] =  df_leak['target_adj'].values
df_sub = df_sub.reset_index()
df_sub.to_csv(DATA.replace('.csv','_adj.csv'),index=False)
df_sub.head()
```
